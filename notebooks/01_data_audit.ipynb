```python
# Leap Year Validation

years = sorted(data["time"].dt.year.unique())

print("Leap Year Validation\n")

for year in years:
    expected = 8784 if pd.Timestamp(year=year, month=12, day=31).is_leap_year else 8760

    actual = (
        data.loc[data["time"].dt.year == year, "time"]
        .nunique()
    )

    print(
        f"{year} | Expected Hours: {expected:,} | "
        f"Actual Hours: {actual:,} | "
        f"{'PASS' if expected == actual else 'CHECK'}"
    )
```

```python
# Expected vs Actual Records

num_cities = data["city"].nunique()

expected_records = 0

for year in years:
    hours = 8784 if pd.Timestamp(year=year, month=12, day=31).is_leap_year else 8760
    expected_records += hours * num_cities

actual_records = len(data)

print(f"Cities            : {num_cities}")
print(f"Expected Records  : {expected_records:,}")
print(f"Actual Records    : {actual_records:,}")
print(f"Difference        : {expected_records - actual_records:,}")
```

### Value Range Validation

```python
# Validation Rules

validation_rules = {
    "dni": (0, None),
    "dhi": (0, None),
    "ghi": (0, None),
    "temperature_2m": (-50, 60),
    "relative_humidity_2m": (0, 100),
    "rain": (0, None),
    "precipitation": (0, None),
    "wind_speed_10m": (0, None),
    "wind_direction_10m": (0, 360),
    "wind_direction_100m": (0, 360),
    "wind_gusts_10m": (0, None),
    "wind_speed_100m": (0, None),
    "cloud_cover": (0, 100),
    "sunshine_duration": (0, 3600),
    "snowfall": (0, None),
    "snow_depth": (0, None),
    "pressure_msl": (800, 1100),
    "surface_pressure": (500, 1100),
    "vapour_pressure_deficit": (0, None),
    "latitude": (-90, 90),
    "longitude": (-180, 180)
}
```

```python
results = []

for column, (lower, upper) in validation_rules.items():

    invalid = pd.Series(False, index=data.index)

    if lower is not None:
        invalid |= data[column] < lower

    if upper is not None:
        invalid |= data[column] > upper

    results.append({
        "Variable": column,
        "Minimum": data[column].min(),
        "Maximum": data[column].max(),
        "Invalid Records": invalid.sum(),
        "Status": "✅ Pass" if invalid.sum() == 0 else "❌ Fail"
    })

range_validation = pd.DataFrame(results)

display(range_validation)
```

```python
display(
    range_validation[
        range_validation["Invalid Records"] > 0
    ]
)
```

```python
summary = pd.DataFrame({

    "Metric":[
        "Variables Checked",
        "Variables Passed",
        "Variables Failed",
        "Total Invalid Records"
    ],

    "Value":[
        len(validation_rules),
        (range_validation["Status"]=="✅ Pass").sum(),
        (range_validation["Status"]=="❌ Fail").sum(),
        range_validation["Invalid Records"].sum()
    ]

})

display(summary)
```

### Outlier Detection

```python
# Numerical Variables

numerical_columns = data.select_dtypes(include=["number"]).columns.tolist()

# Exclude metadata columns
exclude_columns = [
    "rank",
    "tier",
    "latitude",
    "longitude"
]

numerical_columns = [
    col for col in numerical_columns
    if col not in exclude_columns
]

print(f"Variables Checked: {len(numerical_columns)}")
```

```python
# IQR Outlier Detection

outlier_results = []

for column in numerical_columns:

    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = (
        (data[column] < lower) |
        (data[column] > upper)
    )

    outlier_results.append({
        "Variable": column,
        "Lower Bound": round(lower, 2),
        "Upper Bound": round(upper, 2),
        "Outliers": outliers.sum(),
        "Outlier (%)": round(outliers.mean() * 100, 2)
    })

outlier_summary = pd.DataFrame(outlier_results)

display(outlier_summary)
```

```python
# Variables with Highest Outlier Percentage

display(
    outlier_summary.sort_values(
        "Outlier (%)",
        ascending=False
    )
)
```

```python
# Outlier Summary

summary = pd.DataFrame({

    "Metric":[
        "Variables Checked",
        "Variables with Outliers",
        "Maximum Outlier Percentage"
    ],

    "Value":[
        len(outlier_summary),
        (outlier_summary["Outliers"] > 0).sum(),
        f"{outlier_summary['Outlier (%)'].max():.2f}%"
    ]

})

display(summary)
```

### Temporal Consistency

```python
# City-wise Temporal Consistency

results = []

for city, group in data.groupby("city"):

    group = group.sort_values("time")

    start = group["time"].min()
    end = group["time"].max()

    expected = pd.date_range(start=start, end=end, freq="h")

    actual = group["time"]

    missing = len(expected.difference(actual))

    duplicate = actual.duplicated().sum()

    continuous = (missing == 0) and (duplicate == 0)

    results.append({
        "City": city,
        "Records": len(group),
        "Start": start,
        "End": end,
        "Missing Hours": missing,
        "Duplicate Hours": duplicate,
        "Continuous": continuous
    })

temporal_consistency = pd.DataFrame(results)

display(temporal_consistency)
```
