### Load all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

### Load all yearly Parquet files into a single DataFrame.

```python
folder = Path("../data/raw/weather")

parquet_files = sorted(folder.rglob("*.parquet"))

data = pd.concat(
    (pd.read_parquet(file) for file in parquet_files),
    ignore_index=True
)

print(f"Files loaded: {len(parquet_files)}")
print(data.shape)
```

### Temporal Coverage

```python
# Observation Period

start_date = data["time"].min()
end_date = data["time"].max()

print(f"Observation Start : {start_date}")
print(f"Observation End   : {end_date}")
print(f"Duration          : {end_date - start_date}")
```

```python
# Basic Time Statistics

print(f"Total Records : {len(data):,}")

print(f"Years  : {data['time'].dt.year.nunique()}")
print(f"Months : {data['time'].dt.to_period('M').nunique()}")
print(f"Days   : {data['time'].dt.date.nunique()}")
print(f"Hours  : {data['time'].nunique():,}")
```

```python
# Year Coverage

year_summary = (
    data["time"]
    .dt.year
    .value_counts()
    .sort_index()
    .rename_axis("Year")
    .reset_index(name="Records")
)

display(year_summary)
```

```python
# Month Coverage

month_summary = (
    data["time"]
    .dt.month_name()
    .value_counts()
    .reindex([
        "January","February","March","April","May","June",
        "July","August","September","October","November","December"
    ])
    .rename_axis("Month")
    .reset_index(name="Records")
)

display(month_summary)
```

```python
# Hour Coverage

hourly_summary = (
    data["time"]
    .dt.hour
    .value_counts()
    .sort_index()
    .rename_axis("day")
    .reset_index(name="Records")
)

display(hourly_summary)
```

```python
# Leap Year Validation

years = sorted(data["time"].dt.year.unique())

print("Leap Year Validation\n")

for year in years:
    expected = 8784 if pd.Timestamp(year=year, month=12, day=31).is_leap_year else 8760

    actual = (
        data.loc[data["time"].dt.year == year, "time"]
        .nunique()
    )

    print(
        f"{year} | Expected Hours: {expected:,} | "
        f"Actual Hours: {actual:,} | "
        f"{'PASS' if expected == actual else 'CHECK'}"
    )
```

```python
# Expected vs Actual Records

num_cities = data["city"].nunique()

expected_records = 0

for year in years:
    hours = 8784 if pd.Timestamp(year=year, month=12, day=31).is_leap_year else 8760
    expected_records += hours * num_cities

actual_records = len(data)

print(f"Cities            : {num_cities}")
print(f"Expected Records  : {expected_records:,}")
print(f"Actual Records    : {actual_records:,}")
print(f"Difference        : {expected_records - actual_records:,}")
```

```python
# Duplicate Timestamp Check

duplicates = data.duplicated(subset=["city", "time"]).sum()

print(f"Duplicate (city, time) records : {duplicates:,}")
```

```python
# Missing Timestamp Check

missing_summary = []

for city, group in data.groupby("city"):

    full_range = pd.date_range(
        start=group["time"].min(),
        end=group["time"].max(),
        freq="H"
    )

    missing = len(full_range.difference(group["time"]))

    missing_summary.append({
        "City": city,
        "Missing Hours": missing
    })

missing_df = pd.DataFrame(missing_summary)

display(missing_df.sort_values("Missing Hours", ascending=False))
```

```python
# Temporal Summary

summary = pd.DataFrame({

    "Metric": [

        "Observation Start",
        "Observation End",
        "Years",
        "Months",
        "Days",
        "Unique Hours",
        "Cities",
        "Expected Records",
        "Actual Records",
        "Duplicate Records"

    ],

    "Value": [

        start_date,
        end_date,
        data["time"].dt.year.nunique(),
        data["time"].dt.to_period("M").nunique(),
        data["time"].dt.date.nunique(),
        f"{data['time'].nunique():,}",
        num_cities,
        f"{expected_records:,}",
        f"{actual_records:,}",
        duplicates

    ]

})

display(summary)
```
